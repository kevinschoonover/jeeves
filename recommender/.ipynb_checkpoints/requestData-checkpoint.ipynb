```python
import requests
import datetime
import random
import numpy as np
url_user = "http://192.168.99.100/api/v1/accounts"
url_visit = "http://192.168.99.100/api/v1/visits"
url_order = "http://192.168.99.100/api/v1/orders"
url_menuitem = "http://192.168.99.100/api/v1/menuitems"

def delete_all(url):
    import requests
    r = requests.get(url)
    for i in range(len(r.json())):
        requests.delete(url+'/'+str(r.json()[i]['id']))
```

```python
delete_all(url_user)
delete_all(url_visit)
delete_all(url_order)
delete_all(url_menuitem)
```

```python
user_names = {}
r = requests.get(url_user)
for i in range(len(r.json())):
    user_names[r.json()[i]['id']] = r.json()[i]['firstName']+"_"+r.json()[i]['lastName']
    print("Users: ", user_names[r.json()[i]['id']], r.json()[i]['id'], r.json()[i]['visits'])

r = requests.get(url_visit)
print("Visits: ", r.json())

r = requests.get(url_order)
print("Orders: ", r.json())

r = requests.get(url_menuitem)
item_id = {}
for i in range(len(r.json())):
    item_id[r.json()[i]['itemName']] = r.json()[i]['id']
    print("Menuitems: ", r.json()[i]['itemName'], r.json()[i]['id'])
c1 = list(item_id.keys())[:5]
c2 = list(item_id.keys())[5:10]
c3 = list(item_id.keys())[10:15]
c4 = list(item_id.keys())[15:20]
c5 = list(item_id.keys())[20:]
categories = [c1,c2,c3,c4,c5]
```
